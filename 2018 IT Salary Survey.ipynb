After cheking for missig values both standard and non-standard,I observed that a lot of data would be lost if i dropped null values altogether, leading to biased results. I therefore decide to backward fill the data "bfill()"

```python
#backward fill
df.bfill(inplace=True)
df.isnull().sum()
```

Still had a very small percentage of missing values left. I decided to drop the remaining null values

```python
#dropping all the bull values in the data set
df.dropna()
df.isnull().sum()
```

The approach didn't work,will try to drop the null values from the specific columns

```python
df.dropna(subset=["YearAgo_salary"])
df.isnull().sum()
```

Resorted to replacing the missimg values with median values since they columns at hand were numerical variables

```python
#getting the mediuan of the columns
mean=round(df["YearAgo_salary"].mean(),2)
mean
```

```python
#replacing null values with mean
df["YearAgo_salary"].fillna(mean,inplace=True)
df.isnull().sum()
```

```python
mean1=round(df["TwoYearsAgo_salary"].mean(),2)
mean1
```

```python
df["TwoYearsAgo_salary"].fillna(mean,inplace=True)
df.isnull().sum()
```

Age Distribution

```python
#histogram
sns.FacetGrid(df,hue='Age',height=10)\
    .map(sns.histplot,'Age')\
        .add_legend()
plt.ylabel("Frequency",fontsize=14)
plt.title('2018 Age Distribution')
```
