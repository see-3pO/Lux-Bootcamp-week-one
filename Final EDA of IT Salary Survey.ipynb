Dataset selection
#I decided to use the 2019 and 2020 datasets since they relatively had the same number and variable types
#2018 dataset EDA was done separately

```python
df1_list=df1[['Timestamp','Age', 'Gender','City', 'Seniority level', 'Position ','Total years of experience','Your main technology / programming language',
              'Number of vacation days','Main language at work','Company type','Сontract duration','Yearly brutto salary (without bonus and stocks) in EUR',
              'Yearly bonus + stocks in EUR',
       ]]
```

```python
df1_list.shape
```

```python
df3_list=df3[['Zeitstempel','Age', 'Gender', 'City', 'Seniority level','Position (without seniority)', 'Years of experience',
       'Your main technology / programming language', 'Number of vacation days', 'Main language at work',
       'Company type', 'Сontract duration','Yearly brutto salary (without bonus and stocks)' , 'Yearly bonus','Yearly stocks' ]]
```

```python
df3_list.shape
```

```python
df1_list.info()
```

#Replacing null values with backward fill method

```python
df1_list.bfill(inplace=True)
df1_list.isnull().sum()
```

```python
df3_list.info()
```

```python
df3_list.bfill(inplace=True)
df3_list.isnull().sum()
```

#Replacing remaining null values for bonus and stock

```python
df3_list.update(df3_list[['Yearly bonus','Yearly stocks']].fillna(0))
```

```python
df3_list.info()
```

```python
df3_list[['Yearly stocks']].describe()
```

#Combining the bonus and stock to one variable called bonus

```python
df3_list['bonus'] = df3_list['Yearly bonus']+df3_list['Yearly stocks']
```

#Dropping combined columns

```python
df3_list.drop(columns=['Yearly bonus','Yearly stocks'],inplace= True)
```

```python
df3_list.info()
```

```python
df1_list.info()
```

```python
df3_list.rename(columns={"Yearly brutto salary (without bonus and stocks)":"Yearly gross salary w/o bonuses&stocks"}, inplace=True)
```

```python
df1_list.rename(columns={"Yearly brutto salary (without bonus and stocks) in EUR":"Yearly gross salary w/o bonuses&stocks"}, inplace=True)
```

#Creating standard and new columns for the new dataset and adding them to the dfs

```python
new_col_names = ['timestamp','age','gender','city','level','position','total_exp','techonology','vacation','language','comp_type','contract_duration','gross_sal','bonus']
```

```python
df1_list.columns = new_col_names
df3_list.columns = new_col_names
```

#Concatenating the dataframes to one 'final_df'

```python
#concatenating the data sets
final_df = pd.concat([df3_list,df1_list],ignore_index=True)
```

Data Visualization

```python
sns.FacetGrid(final_df,hue='age',height=8)\
    .map(sns.histplot,'age')\
        .add_legend()
plt.ylabel("Frequency",fontsize=14)
plt.title('Age Distribution')
```

```python
sns.FacetGrid(final_df,hue='gender',height=7)\
    .map(sns.histplot,'gender')\
        .add_legend()
plt.ylabel("Frequency",fontsize=14)
plt.title('Gender Distribution')
```

```python
final_df.boxplot(column=['age'],by=['level'],figsize = (15,10),rot=45, fontsize=9)
```

```python
final_df['contract_duration'].value_counts().plot.pie()
plt.gca().set_aspect('equal')
```
